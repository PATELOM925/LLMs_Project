```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(temperature=0,model_name='gpt-3.5-turbo-0125')
```

```python
from neo4j import GraphDatabase

#show graph resulting from given query
cypher = "MATCH (s)-[r:!MENTIONS]->(t) RETURN s,r,t LIMIT 50"

def showGraph(cypher: str=cypher):

  #creating neo4j sessions to run our queries
  driver = GraphDatabase.driver(
      uri = os.environ['NEO4J_URI'],
      auth = (os.environ['NEO4J_USER'], os.environ['NEO4J_PASSWORD']))
  session = driver.session()
  widget = GraphWidget(graph = session.run(cypher).graph())
  widget.node_label_mapping = 'id'
  display(widget)
  return(widget)
```

```python
# Query Operation

# Creating Vector Representation
from langchain_openai import OpenAIEmbeddings

vector_index = Neo4jVector.from_existing_graph(
    OpenAIEmbeddings(),
    search_type = 'hybrid',
    node_label = 'Document',
    text_node_properties = ['text'],
    embedding_node_property = 'embedding'
)
```

```python
#prompt

prompt = ChatMessagePromptTemplate.format_messages(
    [
        (
            'system',
            'You are extracting organization and person entities from text,',
        ),
        (
            'human',
            'use the given format to extract information from the following',
            'input: {question}'
        )
    ]
)
```

```python
# Creating Chain and Entities

from langchain_core.pydantic_v1 import BaseModel, Field

class Entities(BaseModel):
    name: List[str] = Field(
        ...,
        description="List of organization and person entities that appear in the text"
    )
```

```python
entity_chain = prompt | llm.with_structured_output(entities=Entities)
```

```python
entity_chain.invoke({'question':'Where was sardar patel born?'}).name
```
